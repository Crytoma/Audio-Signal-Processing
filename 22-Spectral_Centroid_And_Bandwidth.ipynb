```python
import IPython.display as ipd, numpy as np, librosa, matplotlib.pyplot as plt
```

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512
```

```python
debussy_file, sr = librosa.load('audio/debussy.wav')
redhot_file, _ = librosa.load('audio/redhot.wav')
duke_file, _ = librosa.load('audio/duke.wav')
```

```python
sc_debussy = librosa.feature.spectral_centroid(debussy_file, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

sc_redhot = librosa.feature.spectral_centroid(redhot_file, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

sc_duke = librosa.feature.spectral_centroid(duke_file, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
sc_debussy.shape 
```

Visualise spectral centroid across time

```python
time = librosa.frames_to_time(range(len(sc_debussy))) #aka frames
```

```python
plt.figure(figsize=(25, 10))

plt.plot(time, sc_debussy, color='blue')
plt.plot(time, sc_redhot, color='red')
plt.plot(time, sc_duke, color='black')
plt.show()
```

Visualise Bandwidth 

```python
ban_debussy = librosa.feature.spectral_bandwidth(debussy_file, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

ban_redhot = librosa.feature.spectral_bandwidth(redhot_file, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

ban_duke = librosa.feature.spectral_bandwidth(duke_file, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]

plt.figure(figsize=(25, 10))

plt.plot(time, ban_debussy, color='blue')
plt.plot(time, ban_redhot, color='red')
plt.plot(time, ban_duke, color='black')
plt.show()
```
